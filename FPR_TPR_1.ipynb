### Строим модель

```python
# формируем данные
features_columns = [c for c in data.columns if c != 'sex']
X = data[features_columns]
X.head(3)
```

```python
# целевая переменная
y = data['sex']
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le = LabelEncoder()
ohe = OneHotEncoder()
```

```python
ft = ohe.fit_transform(data[['nationality', 'height', 'weight', 'sport', 'gold', 'silver', 'bronze']])
```

```python
X = data[['height', 'weight', 'gold', 'silver', 'bronze']]
X.head(4)
```

```python
le.fit( data['sex'] )
```

```python
y = pd.Series( data = le.transform( data['sex'] ) )
y.head()
```

```python
model = LogisticRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_curve
```

```python
predictions = model.predict_proba(X_test)
```

```python
fpr, tpr, thres = roc_curve(y_test, predictions[:, 1])
print('fpr:', fpr, '\ntpr:', tpr, '\nthres:', thres)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot( fpr, tpr )
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, predictions[:, 1])
```

Домашние задание

```python
tpr_1 = []
fpr_1 = []
precision = []

predictions = model.predict_proba(X_test)
predictions_1 = model.predict_proba(X_test)
pred = predictions_1[:, 1] # требует только вероятность для положительного случая
pred.sort()
for i in pred:
    
    tp = 0 # True positive
    fp = 0 # False positive
    fn = 0 # False negative
    tn = 0 # True negative
    
    for predicted_prob, actual in zip(predictions[:, 1], y_test):
        
        if predicted_prob >= i:
            predicted = 1
        else:
            predicted = 0
            
        if predicted == 1:
            if actual == 1:
                tp += 1
            else:
                fp += 1
        
        else:
            if actual == 1:
                fn += 1
            else:
                tn += 1
        
    if fn == 0:
        None
    else:
        fpr_1.append(fp / (fp + tn))
        tpr_1.append(tp / (tp + fn))
        precision.append(tp / (tp + fp))
```

```python
print('fpr_1 - false positive rate:', fpr_1)
print('tpr_1 - true positive rate:', tpr_1)
print(len(fpr_1))
print(len(tpr_1))
```

```python
from matplotlib import pyplot as plt
plt.plot(fpr, tpr)
plt.plot(fpr_1, tpr_1)
```

Precision and recall

```python
recall = tpr_1
print('Precision:', precision)
```

```python
plt.plot(recall, precision)
```
