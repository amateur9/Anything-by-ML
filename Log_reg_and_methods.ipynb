```python
import numpy as np
import pandas as pd 
import scipy.stats 
import matplotlib.pyplot as plt  
import math
```

```python
N = 100
X = np.random.uniform(low = 0, high = 100, size = N)
Y = 30 * X + np.random.normal(scale=5, size=N)
plt.scatter(X, Y)
plt.show()
```

```python
print('Градиентный спуск')
w = 1
EPOCHS = 20
step = 0.0001
preds = []

for _ in range(EPOCHS):
    value_w = 0
    value_loss = 0
    for i in range(len(X)):
        value_w += (w * X[i] - Y[i]) * X[i] / len(X)
        value_loss += (w * X[i] - Y[i]) ** 2 / (2 * len(X))
    w = w - step * value_w
    print("w:", w, "cost:", value_loss)
    predictions = w * X
    preds.append(predictions)
plt.scatter(X, Y, color='blue')
plt.scatter(X, preds[19], color='red')
plt.show()
```

```python
print('Нестеров')
w = 1
EPOCHS = 38
gamma = 0.9999
step_w = 0
alfa = 0.0001
preds = []

for _ in range(EPOCHS):
    value_w = 0
    value_loss = 0
    for i in range(len(X)):
        value_w += (w * (X[i] - step_w) - Y[i]) * (X[i] - step_w) / len(X)
        value_loss += (w * (X[i] - step_w) - Y[i]) ** 2 / (2 * len(X))
    step_w = gamma * step_w + (1 - gamma) * value_w
    w = w - step_w 
    print("w:", w, "cost:", value_loss)
    
    predictions = w * X
    preds.append(predictions)
    
plt.scatter(X, Y, color = 'blue')
plt.scatter(X, preds[37], color = 'red')
plt.show()
```

```python
print('RMSProp')
w = 1
EPOCHS = 50
gamma = 0.001
step_w = 0
epsilon = 0.000001
preds = []

for _ in range(EPOCHS):
    value_w = 0
    value_w_2 = 0
    value_loss = 0
    for i in range(len(X)):
        value_w += (w * X[i] - Y[i]) * X[i] / len(X) 
        value_w_2 += ((w * X[i] - Y[i]) * X[i]) ** 2 / len(X)
        value_loss += (w * X[i] - Y[i]) ** 2 / (2 * len(X))
    step_w = gamma * step_w + (1 - gamma) * value_w_2
    w = w - (1 - gamma) * value_w / math.sqrt(step_w + epsilon)
    print("w:", w, "cost:", value_loss)
    predictions = w * X
    preds.append(predictions)
plt.scatter(X, Y, color='blue')
plt.scatter(X, preds[EPOCHS - 1], color='red')
plt.show()
```
