Задача на расчёт налога ИП при упрощенной системе налогооблажения

```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.linear_model import LinearRegression
```

```python
amount_businessmen = 500

revenue_1 = np.random.choice(460000, amount_businessmen) + 94000
revenue_2 = np.random.choice(580000, amount_businessmen) + 16000
revenue_3 = np.random.choice(950000, amount_businessmen) + 57000
revenue_4 = np.random.choice(760000, amount_businessmen) + 79000

ins_payment_3 = np.random.choice(47000, amount_businessmen) + 3000
ins_payment_6 = np.random.choice(70000, amount_businessmen) + 5300
ins_payment_9 = np.random.choice(94000, amount_businessmen) + 8200
ins_payment_12 = np.random.choice(79000, amount_businessmen) + 2000

prepaid_1 = 0.06 * revenue_1 - ins_payment_3
# for i in range(len(prepaid_1)):
#     if prepaid_1[i] < 0:
#         prepaid_1[i] = 0
        
prepaid_2 = 0.06 * (revenue_1 + revenue_2) - ins_payment_6 - prepaid_1
# for i in range(len(prepaid_2)):
#     if prepaid_2[i] < 0:
#         prepaid_2[i] = 0
        
prepaid_3 = 0.06 * (revenue_1 + revenue_2 + revenue_3) - ins_payment_9 - prepaid_1 - prepaid_2
# for i in range(len(prepaid_3)):
#     if prepaid_3[i] < 0:
#         prepaid_3[i] = 0
        
result = (0.06 * (revenue_1*revenue_1 + revenue_2 + revenue_3 + revenue_4) - ins_payment_12 -prepaid_1 - prepaid_2 - prepaid_3)
# for i in range(len(result)):
#     if result[i] < 0:
#         result[i] = 0

data = pd.DataFrame({'revenue_1': revenue_1, 'revenue_2': revenue_2, 'revenue_3': revenue_3, 'revenue_4': revenue_4,
                    'ins_payment_3': ins_payment_3, 'ins_payment_6': ins_payment_6, 'ins_payment_9': ins_payment_9, 
                    'ins_payment_12': ins_payment_12, 'prepaid_1': prepaid_1, 'prepaid_2': prepaid_2, 
                    'prepaid_3': prepaid_3, 'result': result})
data.head(5)
```

```python
from sklearn.metrics import mean_absolute_error

X = data[['revenue_1', 'revenue_2', 'revenue_3', 'revenue_4']]
y = data['result']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_)) # коэф k
print('Bias: {}'.format(reg.intercept_)) # коэф m

pred_values = reg.predict(data[['revenue_1', 'revenue_2', 'revenue_3', 'revenue_4']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```
